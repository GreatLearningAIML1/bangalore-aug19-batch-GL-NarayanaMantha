# Step - 4.a
## Do charges of people who smoke differ significantly from the people who don't

```python
sns.catplot(x="smoker", y="charges",data=df)
```

## Observation - As we can see from the above plot, there is a significant difference between the charges for a smoker and non-smoker
## We will try to verify this in statistical way using hypothesis testing

```python
from scipy.stats import levene, shapiro, mannwhitneyu, chi2_contingency, chisquare
```

```python
smoker = df["charges"][df["smoker"] == 'yes']
non_smoker = df["charges"][df["smoker"] == 'no']

#For doing hypothesis testing, we need to find a correct test which can be used on the above data. This can be decided based
# 1. on the similarity of variance between the above 2 groups
# 2. also on the normality distribution of the above 2 groups

# We will check if the variance between the 2 groups is similar or not by using Levene's test
stat, p_val = levene(smoker, non_smoker)
if(p_val > 0.05):
    print("Both groups has equal variance")
else:
    print("Both groups have different variance")

#Now to check whether the distribution is normal or not, we will use Shapiro test on both the groups separately

#checking the distribution for smoker group
stat, p_val = shapiro(smoker)
if(p_val > 0.05):
    print("Smoker group is normally disributed")
else:
    print("Smoker group distribution is not normal")

#checking the distribution for smoker group
stat, p_val = shapiro(non_smoker)
if(p_val > 0.05):
    print("Non Smoker group is normally disributed")
else:
    print("Non Smoker group distribution is not normal")
```

```python
# As we can see from the above output,
# 1. the variance between the 2 groups is significantly different
# 2. and also the distributions for both groups are not normal.

# In this case, we will use mannwhitneyu to determine if the charges are different for these 2 groups(whose variance is different
# and the distribution is not normal)

# Null Hypothesis: There is no significant difference in insurance charges between these groups
# Alternate Hypothesis: There is significant difference in insurance charges between these groups
u, p_val = mannwhitneyu(smoker, non_smoker)
print("p_val = ", p_val)
if(p_val > 0.05):
    print("There is no significant difference between smoker and non-smoker insurance charge")
else:
    print("There is significant difference between smoker and non-smoker insurance charge")
```

### Conclusion - Since the p-value is less than level of siginificance, we are rejecting Null Hypothesis. It means that there is a significant difference in insurance charges between smoker and non-smoker in this data set.

```python
male_bmi = df["bmi"][df["sex"] == 'male']
female_bmi = df["bmi"][df["sex"] == 'female']

#For doing hypothesis testing, we need to find a correct test which can be used on the above data. This can be decided based
# 1. on the similarity of variance between the above 2 groups
# 2. also on the normality distribution of the above 2 groups

# We will check if the variance between the 2 groups is similar or not by using Levene's test
stat, p_val = levene(male_bmi, female_bmi)
if(p_val > 0.05):
    print("Both groups has equal variance")
else:
    print("Both groups have different variance")

#Now to check whether the distribution is normal or not, we will use Shapiro test on both the groups separately

#checking the distribution for male_bmi group
stat, p_val = shapiro(male_bmi)
if(p_val > 0.05):
    print("male_bmi group is normally disributed")
else:
    print("male_bmi group distribution is not normal")

#checking the distribution for smoker group
stat, p_val = shapiro(female_bmi)
if(p_val > 0.05):
    print("female_bmi group is normally disributed")
else:
    print("female_bmi group distribution is not normal")
```

```python
# As we can see from the above output,
# 1. Both groups have equal variance
# 2. But the distributions for both groups are not normal.

# In this case, we will use mannwhitneyu to determine if the charges are different for these 2 groups(whose variance is equal
# and the distribution is not normal)

# Null Hypothesis: There is no significant difference in bmi between male and female
# Alternate Hypothesis: There is significant difference in bmi between male and female
u, p_val = mannwhitneyu(male_bmi, female_bmi)
print("p_val = ", p_val)
if(p_val > 0.05):
    print("There is no significant difference in bmi between male and female")
else:
    print("There is significant difference in bmi between male and female")
```

### Conclusion - Since the p-value is greater than level of siginificance, we are not rejecting Null Hypothesis. It means that there is no significant difference in bmi between male and female in this data set. 

# Step - 4.c
## Is the proportion of smokers significantly different in different genders

```python
clarity_color_table = pd.crosstab(index=df["sex"], 
                          columns=df["smoker"])

clarity_color_table
```

```python
clarity_color_table.plot(kind="bar", 
                 figsize=(8,8),
                 stacked=True)
```

### Observation - As we can see from the above plot, the proportions of smokers significantly differ in both male and female genders.

## We will try to verify this in statistical way using hypothesis testing

```python
# we will use a chi2 test for this, as it involves categorical variable
# we will try to check the proortion of smokers in males and females separately

male_smoker = df["smoker"][df["sex"] == 'male']
female_smoker = df["smoker"][df["sex"] == 'female']

#Null Hypothesis: Male smoker & non smoker proportions are same
#Alternate Hypothesis: Male smoker & non smoker proportions are not same
stat, p_val = chisquare(male_smoker.value_counts())
print("p_val for male_smokers is ", p_val)

if( p_val > 0.05):
    print("Male smoker & non smoker proportions are same")
else:
    print("Male smoker & non smoker proportions differ")

#Null Hypothesis: Female smoker & non smoker proportions are same
#Alternate Hypothesis: Female smoker & non smoker proportions are not same
stat, p_val = chisquare(female_smoker.value_counts())
print("p_val for female_smokers is ", p_val)

if( p_val > 0.05):
    print("Female smoker & non smoker proportions are same")
else:
    print("Female smoker & non smoker proportions differ")
```

### Conclusion - Since the p-value is far less than the level of siginificance, we are rejecting Null Hypothesis. It means that the proportions of smokers significantly differ in both male and female genders.
